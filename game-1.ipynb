```python
#Python code for "Rock,Paper,Scissors" game
import tkinter as tk
import tkinter.messagebox as mb
import random
#========setting=========
root=tk.Tk()
root.title("Rock,Paper,Scissors")
root.geometry('200x320')
root.resizable(width=False,height=False)
root.configure(bg="lightcoral")
#=======Frame=============
f1=tk.Frame(root,width=200,height=50,bg="lightcoral")
f1.pack()
f2=tk.Frame(root,width=200,height=50,bg="lightcoral")
f2.pack()
f3=tk.Frame(root,width=200,height=50,bg="lightcoral")
f3.pack()
f4=tk.Frame(root,width=200,height=50,bg="lightcoral")
f4.pack()
f5=tk.Frame(root,width=200,height=50,bg="lightcoral")
f5.pack()
f6=tk.Frame(root,width=200,height=50,bg="lightcoral")
f6.pack()
f7=tk.Frame(root,width=200,height=50,bg="lightcoral")
f7.pack()
f8=tk.Frame(root,width=200,height=50,bg="lightcoral")
f8.pack()
#===========Button==========
b1=tk.Button(f1,text="Rock",width=20,font="arial 12",command=lambda:rock())
b1.pack(padx=25,pady=5)
b2=tk.Button(f2,text="Paper",width=20,font="arial 12",command=lambda:paper())
b2.pack(padx=25,pady=5)
b3=tk.Button(f3,text="Scissors",width=20,font="arial 12",command=lambda:scissors())
b3.pack(padx=25,pady=5)
b4=tk.Button(f5,text="",width=20,font="arial 12")
b4.pack(padx=25,pady=5)
b5=tk.Button(f7,text="",width=20,font="arial 12")
b5.pack(padx=25,pady=5)
b6=tk.Button(f8,text="clear",width=5,font="arial 10",command=lambda:clear())
b6.pack(side="left",padx=25,pady=20)
b7=tk.Button(f8,text="creator",width=5,font="arial 10",command=lambda:creator())
b7.pack(side="left",padx=25,pady=20)
#=========Label===========
L1=tk.Label(f4,text="Computer move:",font="arial 12 bold",bg="lightcoral",fg="lightblue")
L1.pack()
L2=tk.Label(f6,text="Result:",font="arial 12 bold",bg="lightcoral",fg="lightblue")
L2.pack()
#============Function=======
def creator():
    mb.showinfo("creator","This game has been created by Atena Pakzad with Python")
def rock():
    z=random.randint(0,2)
    if z==0:
        b4["text"]="Rock"
        b5["text"]="tie"
    elif z==1:
        b4["text"]="Paper"
        b5["text"]="You lost!"
    else:
        b4["text"]="Scissors"
        b5["text"]="You won!"
        
def paper():
    z=random.randint(0,2)
    if z==0:
        b4["text"]="Rock"
        b5["text"]="You won!"
    elif z==1:
        b4["text"]="Paper"
        b5["text"]="tie"
    else:
        b4["text"]="Scissors"
        b5["text"]="You lost!"
def scissors():
    z=random.randint(0,2)
    if z==0:
        b4["text"]="Rock"
        b5["text"]="You lost!"
    elif z==1:
        b4["text"]="Paper"
        b5["text"]="You wone!"
    else:
        b4["text"]="Scissors"
        b5["text"]="tie!"
def clear():
    b4["text"]=""
    b5["text"]=""               
#========Run==============
root.mainloop()
```
